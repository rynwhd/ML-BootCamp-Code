```python
 # Feature Engineering the Data
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df['date']
```

```python
df['year']=df['date'].apply(lambda x:x.year)
```

```python
df['month']=df['date'].apply(lambda x:x.month)
```

```python
sns.boxplot(x='year',y='price',data=df)
```

```python
df.drop('date',axis=1,inplace=True)
```

```python
df.drop('zipcode',axis=1,inplace=True) # we are dropping zipcode as it seems to be categorical data rather than 
#continuous data. We could categorise it through one-hot-encoding but that takes up too many columns
```

```python
#Model Creation and Processing
```

```python
X=df.drop(['price'],axis=1).values
```

```python
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.fit_transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(19,activation='relu'))#We chose 19 as incoming data features are 19 and we want to continue with it
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(X_train,y_train,verbose=1,validation_data=(X_test,y_test),batch_size=128,epochs=500) # this validation is used by the network to keep 
#track with the test data, it doesn't effect out training process which is essentially done by the training data
#BATCH_SIZE is feeding the data in power's of 2 so that we don't overfit the data
#
```

```python
# Model Evaluation 
```

```python
losses=pd.DataFrame(model.history.history) #here we have val_loss which is the loss on the test data
```

```python
losses.plot() #here if we see that the val_loss is spiking after a certain point, we can say that it is overfitting 
#on our data. Which can be stopped by early stopping
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
predictions=model.predict(X_test)
```

```python
mean_absolute_error(y_test,predictions)
```

```python
mean_squared_error(y_test,predictions)**0.5
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(x=y_test,y=predictions)
plt.plot(y_test,y_test,'y')
```
