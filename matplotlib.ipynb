```python
import matplotlib.pyplot as plt
#allows seeing plots in the jupyter nb
%matplotlib inline 
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional method 
plt.plot(x,y)
```

```python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Graph')
plt.plot(x,y)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method

fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8]) #gives axis left, bottom , width , height
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])#this is basically percent between 0 and 1 which give us portion of the chart it will cover
axes1.plot(x,y,'g')
axes1.set_title('Larger Plot')
axes2=fig.add_axes([.2,.2,.2,.2])
axes2.plot(y,x,'y')
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
axes[0][0].plot(x,y,'r')
axes[0][1].plot(y,x,'b')
axes[0][2].plot(x,y,'g')
```

```python
# Figure Size and DPI
fig=plt.figure(figsize=(8,2))
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

```python
fig,axes=plt.subplots(figsize=(8,2),nrows=2,ncols=1)
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#save a figure

fig.savefig('myfigure.jpg',dpi=200)
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,'r',label='Square')
axes.plot(x,x**3,'b',label='Cube')#this label gives legend the name for color

axes.legend()
```

```python
#Customising the plots
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='orange',linewidth=5,alpha=.5,linestyle='--')
```

```python
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',linewidth=3,linestyle='-',marker='.',markersize=15)#
```
